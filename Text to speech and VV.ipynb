```python
import tkinter as tk
from tkinter import*
from tkinter import filedialog
from tkinter.ttk import Combobox
import pyttsx3
import os
from pyttsx3.voice import Voice
import speech_recognition as sr

root= Tk()
root.title("Manjot Kaur")
root.geometry("900x450")
root.resizable(False,False)
root.configure(bg="#305065")
        
    
def one():
        root.destroy()
        root1= Tk()
        root1.title("Speech to text")
        root1.geometry("900x450")
        root1.resizable(False,False)
        root1.configure(bg="#305065")

        engine = pyttsx3.init()
        def speaknow():
            r = sr.Recognizer()
            with sr.Microphone() as source:
                audio = r.listen(source)
                a = r.recognize_google(audio)
            text_area.insert(END, a)    


        #icon
        image_icon = PhotoImage(file="speak.png")
        root1.iconphoto(False,image_icon)

        #Top Frame
        Top_frame=Frame(root1,bg="white",width=990,height=100)
        Top_frame.place(x=0,y=0)

        Logo= PhotoImage(file="speakerlogo.png")
        L1=Label(Top_frame,image=Logo,bg="white")
        L1.place(x=10,y=-5)

        Label(Top_frame,text="SPEECH TO TEXT",font="arial 20 bold",bg="white",fg="black").place(x=200,y=30)


        text_area = Text(root1,font="Robote 20",bg="white",relief= GROOVE , wrap = WORD)
        text_area.insert(INSERT,"Say Something:\n")
        text_area.place(x=10,y=150,width=500,height=250)



        imageicon=PhotoImage(file="speak.png")
        btn=Button(root1,text="Say",compound='left',image=imageicon,width=160,height=80,font="arial 14 bold",command=speaknow)
        btn.place(x=630,y=230)
        root1.mainloop()


def two():
        root.destroy()
        root2= Tk()
        root2.title("Text to speech")
        root2.geometry("900x450")
        root2.resizable(False,False)
        root2.configure(bg="#305065")

        engine = pyttsx3.init()
        def speaknow():
            text=text_area.get(1.0,END)
            gender=gender_combobox.get()
            speed=speed_combobox.get()
            voices=engine.getProperty("voices")

            def setvoice():
                if(gender == 'Male'):
                    engine.setProperty("voice",voices[0].id)
                    engine.say(text)
                    engine.runAndWait()
                else:
                    engine.setProperty("voice",voices[1].id)
                    engine.say(text)
                    engine.runAndWait()
            if(text):
                if(speed=="Fast"):
                    engine.setProperty("rate",250)
                    setvoice()
                elif(speed=="Normal"):
                    engine.setProperty("rate",150)
                    setvoice()
                else:
                    engine.setProperty("rate",60)
                    setvoice()


        def download():
            text=text_area.get(1.0,END)
            gender=gender_combobox.get()
            speed=speed_combobox.get()
            voices=engine.getProperty("voices")

            def setvoice():
                if(gender == 'Male'):
                    engine.setProperty("voice",voices[0].id)
                    path=filedialog.askdirectory()
                    os.chdir(path)
                    engine.save_to_file(text,"text.mp3")
                    engine.runAndWait()
                else:
                    engine.setProperty("voice",voices[1].id)
                    path=filedialog.askdirectory()
                    os.chdir(path)
                    engine.save_to_file(text,"text.mp3")
                    engine.runAndWait()
            if(text):
                if(speed=="Fast"):
                    engine.setProperty("rate",250)
                    setvoice()
                elif(speed=="Normal"):
                    engine.setProperty("rate",150)
                    setvoice()
                else:
                    engine.setProperty("rate",60)
                    setvoice()



        #icon
        image_icon = PhotoImage(file="speak.png")
        root2.iconphoto(False,image_icon)

        #Top Frame
        Top_frame=Frame(root2,bg="white",width=990,height=100)
        Top_frame.place(x=0,y=0)

        Logo= PhotoImage(file="speakerlogo.png")
        L1=Label(Top_frame,image=Logo,bg="white")
        L1.place(x=10,y=-5)

        Label(Top_frame,text="TEXT TO SPEECH",font="arial 20 bold",bg="white",fg="black").place(x=200,y=30)

        text_area = Text(root2,font="Robote 20",bg="white",relief= GROOVE , wrap = WORD)
        text_area.place(x=10,y=150,width=500,height=250)

        Label(root2,text="VOICE",font ='arial 15 bold',bg="#305065",fg="white").place(x=580,y=160)
        Label(root2,text="SPEED",font ='arial 15 bold',bg="#305065",fg="white").place(x=760,y=160)

        gender_combobox = Combobox(root2, values=['Male','Female'],font="arial 14",state="r",width=10)
        gender_combobox.place(x=550,y=200)
        gender_combobox.set('Male')

        speed_combobox = Combobox(root2, values=['Fast','Normal','Slow'],font="arial 14",state="r",width=10)
        speed_combobox.place(x=730,y=200)
        speed_combobox.set('Normal')

        imageicon=PhotoImage(file="speak.png")
        btn=Button(root2,text="Speech",compound='left',image=imageicon,width=140,height=50,font="arial 14 bold",command=speaknow)
        btn.place(x=550,y=280)

        imageicon2=PhotoImage(file="download1.png")
        btn1=Button(root2,text="Save",compound='left',image=imageicon2,width=140,height=50,bg="#39c790",font="arial 14 bold",command=download)
        btn1.place(x=730,y=280)
        root2.mainloop()


Top_frame=Frame(root,bg="white",width=990,height=100)
Top_frame.place(x=0,y=0)

Label(Top_frame,text="Text to speech and Speech To Text Converter",font="arial 20 bold",bg="white",fg="black").place(x=140,y=30)
    
B1 = Button(root,text='Speech To Text',width=20,height=2,font="arial 20 bold",command=one)
B1.place(x=280,y=150)

B2 = Button(root,text='Text To Speech',width=20,height=2,font="arial 20 bold",command=two)
B2.place(x=280,y=300)

root.mainloop()
```
